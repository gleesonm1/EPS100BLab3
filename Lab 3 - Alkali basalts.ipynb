# Origin, evolution, and characteristics of alkali basalts

Alkali basalts are one of two general types of basaltic magma, the other type is tholeiitic basalt. Basalts erupted at mid-ocean ridges are usually tholeiitic, basalts erupted at intraplate island volcanoes, like Jan Mayan (N. Atlantic; we'll be looking at rocks from here later on in the lab), are generally alkali basalt - although tholeiites are observed at volcanoes the hottest and most active plumes (e.g., Hawai'i).

Alkali basalt is distinguished from tholeiitic basalts in several ways:

<ol>
    <li>Alkali basalts have more Na$_2$O and K$_2$O (have a look at any TAS diagram with the alkali-subalkali/tholeiitic divide).
    <li>Alkali basalts contain normative Nepheline and never contain normative orthopyroxene or quartz (see the basalt tetrahedron).
    <li>Alkali basalts do not contain orthopyroxene. The common phenocrsts are olivine, clinopyroxene and plagioclase feldspar (also the common phenocrysts of tholeiites, but tholeiites sometimes contain orthopyroxene).
    <li>Alkali basalts typically have higher concentrations of incompatible trace elements. 
</ol>

Alkali basalt is believed to form by mantle melting taking place at deeper levels beneath the surface (>60 - 100 km) than tholeiitic basalts (25 - 60 km), and under conditions where the amount of mantle melting is smaller (~1 - 4%) than for tholeiitic basalts (>5%).

The purpose of this lab is to first consider the formation of alkali basalts, before getting some experience of what these basalts look like. We'll also think about what happens as these basalts crystallise, highlighting that minerals that crystallise from basalt typically contain less SiO$_2$ than the magmas they crystallise from. Consequently, as magma crystallise and the crystals settle out because of gravity or other processes, the magma that is left over is richer in silica (typically referred to as fractional crystallisation or magmatic differentiation). Therefore, in any volcanic area, there is typically a sequence of lava compositions that represent some original magma type (usually basalt) that has crystallised or differentiated to different degrees prior to eruption. This sequence of lava compositions is called a diffferentiation series. In some sequences, one can relate the magma compositions directly to the compositions and the amount of the minerals that have started and separated out prior to eruption.

## Understanding the basalt tetrahedron

The composition of all basaltic magmas can be described by calculating their normative mineral assemblage. All basalts contain normative plagioclase (An) and clinopyroxene (Di) but are subdivided based on which pair of nepheline, olivine, orthopyroxene, and quartz are also present in their normative mineral assemblage. Basalts containing nepheline by default also contain normative olivine and are classificed as alkali basalts (silica undersaturated). Basalts containing normative quartz must also contain normative orthopyroxene and are characterised as quartz tholeiites (silica oversaturated). Finally, basalts that have neither quartz or nepheline in their **norm** are characterised by normative olivine and orthopyroxene and are classified as olivine tholeiites (silica saturated).

Lets think about what is actually happening during mantle melting.

At the point a magma is formed in the mantle it is in chemical equilibrium with the residue of the solid mantle from which it formed. As the mantle is dominated by the minerals olivine, orthopyroxene, and clinopyroxene, magmas forming in the mantle must be in equilibrium with these three phases at the pressure of melt formation. Using this logic, we can use the basalt tetrahedron (and resulting pseudoternary phase diagrams) to identify the composition of melts formed during melting of a lherzolitic mantle.

However, there is a complication. The phase diagrams that you saw in lab one all assumed a fixed pressure (i.e., they were isobaric). From last week's lab, we know that this assumption doesn't hold for mantle melting. Nevertheless, by assessing phase diagrams calibrated at different pressures, we can learn a lot about why the pressure of melting appears to influence the generation of alkali and tholeiitic basalts.
